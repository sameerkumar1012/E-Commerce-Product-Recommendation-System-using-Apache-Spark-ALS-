```python
df = spark.read.csv("events.csv", header=True, inferSchema=True)
df.show(5)
```

```python
df_small = df.limit(200000)
df_small.show(5)
```

```python
from pyspark.sql.functions import col, when

ratings = df_small.withColumn(
    "rating",
    when(col("event") == "view", 1)
    .when(col("event") == "addtocart", 2)
    .when(col("event") == "transaction", 3)
)

ratings = ratings.select(
    col("visitorid").alias("userId"),
    col("itemid").alias("itemId"),
    col("rating")
).dropna()
```

```python
ratings = ratings.withColumn("userId", col("userId").cast("int")) \
                 .withColumn("itemId", col("itemId").cast("int")) \
                 .withColumn("rating", col("rating").cast("int"))

ratings.show(5)
```

```python
train, test = ratings.randomSplit([0.8, 0.2], seed=42)
```

```python
from pyspark.ml.recommendation import ALS

als = ALS(
    maxIter=5,
    rank=10,
    regParam=0.1,
    userCol="userId",
    itemCol="itemId",
    ratingCol="rating",
    coldStartStrategy="drop"
)

model = als.fit(train)

print("MODEL TRAINED SUCCESSFULLY!")
```

```python
from pyspark.ml.evaluation import RegressionEvaluator

predictions = model.transform(test)

evaluator = RegressionEvaluator(
    metricName="rmse",
    labelCol="rating",
    predictionCol="prediction"
)

rmse = evaluator.evaluate(predictions)
print("RMSE =", rmse)
```
